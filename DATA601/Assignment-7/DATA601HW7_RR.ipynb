1.a Create a class with the name "<code>rectangle</code>", which accepts 4 numbers (*a, b, x1, y1*) to create a rectangle object. See the image below. *a* and *b* are the side lenghts, and *x1* and *y1* are the *x-* and *y-*coordinates of the lower left corner

image.png

```python
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import numpy as np
```

1.b Write a method which calculates and returns the area of this rectangle.  
1.c Write a method which calculates and returns the perimeter of this rectangle.  
1.d Write a method to plot this rectangle.  
1.e Write a method to shift this rectangle by (delta_x, delta_y), e.g. new x1 = x1+delta_x,new y1 = y1+delta_y.  

```python
class rectangle:
    #1.a          
    # define a constructor

    def __init__(self, a, b, x1, y1):
        self.a = a
        self.b = b
        self.x1 = x1
        self.y1 = y1
        
    #1.b      
    # method to calculate area of rectangle
    def area(self):
        return self.a * self.b
    
    #1.c
    # method to calculate perimeter of rectangle
    def perimeter(self):
        return float(2 * (self.a+self.b))
    
    #1.d
    # method to define shape of rectangle
    def plot_shape(self):
        fig = plt.figure()
        currentAxis = plt.gca()
        currentAxis.add_patch(Rectangle((self.x1, self.y1), self.a, self.b, fill = None))
        plt.axis('scaled')
        plt.show()
        
    #1.e    
    # method to shift the co-ordinates   
    def shift(self, delta_x, delta_y):
        self.x1 = self.x1+delta_x
        self.y1 = self.y1+delta_y
        
    
```

```python
# create a new rectangle object
r1 = rectangle(5,3,0,0)
```

```python
# get the area
r1.area()
```

```python
# get the perimeter
r1.perimeter()
```

```python
# plot the rectangle
r1.plot_shape()
```

```python
# shift the rectangle
r1.shift(2,2)
```

```python
# plot it and check the shift
r1.plot_shape()
```

```python
# inheriting the rectangle to the square class
class square(rectangle):
    def __init__(self,a,x1,y1):
        super().__init__(a,a,x1,y1)
```

```python
# create a square
s1 = square(4,2,2)
```
